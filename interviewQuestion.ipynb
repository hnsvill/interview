```python
# If this notebook was used as an app, I'd look into widgets,
#   but changing the numbers here in the dict is probably fine for now

# Getting this settled took the most time after loading this into a dataframe
def convertIsoStringDateToWeekdayPriority(isoFormattedDate):
    # Converts to datetime, then returns string weekday to pass to the defined priority (the {dictionary})
    actualWeekday = datetime.date.fromisoformat(isoFormattedDate).strftime("%A")
    datePriorityDefined = {
        'Sunday': 3,
        'Monday': 5,
        'Tuesday': 4,
        'Wednesday': 0,
        'Thursday': 1,
        'Friday': 6,
        'Saturday': 2
    }
    return datePriorityDefined[actualWeekday]

# show how the above function works
print(datetime.date.fromisoformat('2019-12-04').strftime("%A"))
print(convertIsoStringDateToWeekdayPriority('2019-12-04'))
```
